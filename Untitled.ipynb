```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration pour l'affichage
plt.style.use('default')
sns.set_palette("husl")

class ExploratoryDataAnalysis:
    def __init__(self, data_path):
        self.df = pd.read_csv(data_path)
        self.basic_info()
        
    def basic_info(self):
        """Informations de base sur le dataset"""
        print("=" * 50)
        print("üìä ANALYSE EXPLORATOIRE DU DATASET TRANSPORT")
        print("=" * 50)
        
        print(f"üìÅ Shape du dataset: {self.df.shape}")
        print(f"üîç Nombre total de trajets: {len(self.df)}")
        
        print("\nüìã Colonnes disponibles:")
        print(self.df.columns.tolist())
        
        print("\nüîé Aper√ßu des donn√©es:")
        print(self.df.head())
        
        print("\nüìù Informations sur les types de donn√©es:")
        print(self.df.info())
        
        print("\nüßÆ Statistiques descriptives:")
        print(self.df.describe())
    
    def check_missing_values(self):
        """Analyser les valeurs manquantes"""
        print("\n" + "=" * 50)
        print("üîç ANALYSE DES VALEURS MANQUANTES")
        print("=" * 50)
        
        missing = self.df.isnull().sum()
        missing_percent = (missing / len(self.df)) * 100
        
        missing_df = pd.DataFrame({
            'Valeurs manquantes': missing,
            'Pourcentage': missing_percent
        })
        
        print(missing_df[missing_df['Valeurs manquantes'] > 0])
        
        if missing.sum() == 0:
            print("‚úÖ Aucune valeur manquante d√©tect√©e!")
    
    def analyze_transport_types(self):
        """Analyser la r√©partition par type de transport"""
        print("\n" + "=" * 50)
        print("üöó ANALYSE PAR TYPE DE TRANSPORT")
        print("=" * 50)
        
        transport_stats = self.df['type transport'].value_counts()
        print("Nombre de trajets par type de transport:")
        print(transport_stats)
        
        # Pourcentage
        print("\nPourcentage par type de transport:")
        print((transport_stats / len(self.df) * 100).round(2))
        
        # Visualisation
        plt.figure(figsize=(10, 6))
        transport_stats.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen', 'gold'])
        plt.title('R√©partition des trajets par type de transport')
        plt.xlabel('Type de transport')
        plt.ylabel('Nombre de trajets')
        plt.xticks(rotation=45)
        plt.tight_layout()
        plt.show()
    
    def analyze_prices(self):
        """Analyser la distribution des prix"""
        print("\n" + "=" * 50)
        print("üí∞ ANALYSE DES PRIX")
        print("=" * 50)
        
        print("Statistiques des prix (FCFA):")
        print(self.df['prix'].describe())
        
        # Prix par type de transport
        print("\nPrix moyens par type de transport:")
        prix_par_type = self.df.groupby('type transport')['prix'].agg(['mean', 'min', 'max', 'std'])
        print(prix_par_type.round(2))
        
        # Visualisation
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        
        # Boxplot des prix par type
        self.df.boxplot(column='prix', by='type transport', ax=axes[0])
        axes[0].set_title('Distribution des prix par type de transport')
        axes[0].set_ylabel('Prix (FCFA)')
        
        # Histogramme des prix
        self.df['prix'].hist(bins=50, ax=axes[1], alpha=0.7)
        axes[1].set_title('Distribution g√©n√©rale des prix')
        axes[1].set_xlabel('Prix (FCFA)')
        axes[1].set_ylabel('Fr√©quence')
        
        plt.tight_layout()
        plt.show()
    
    def analyze_ratings(self):
        """Analyser les notes de rapidit√© et confort"""
        print("\n" + "=" * 50)
        print("‚≠ê ANALYSE DES NOTES (RAPIDIT√â & CONFORT)")
        print("=" * 50)
        
        # Statistiques des notes
        print("Rapidit√© - Statistiques:")
        print(self.df['rapidite'].describe())
        
        print("\nConfort - Statistiques:")
        print(self.df['confort'].describe())
        
        # Notes par type de transport
        notes_par_type = self.df.groupby('type transport')[['rapidite', 'confort']].mean()
        print("\nNotes moyennes par type de transport:")
        print(notes_par_type.round(2))
        
        # Visualisation
        fig, axes = plt.subplots(1, 2, figsize=(15, 5))
        
        # Rapidit√©
        rapidite_par_type = self.df.groupby('type transport')['rapidite'].mean()
        rapidite_par_type.plot(kind='bar', ax=axes[0], color='lightblue')
        axes[0].set_title('Rapidit√© moyenne par type de transport')
        axes[0].set_ylabel('Rapidit√© (0-10)')
        axes[0].set_ylim(0, 10)
        
        # Confort
        confort_par_type = self.df.groupby('type transport')['confort'].mean()
        confort_par_type.plot(kind='bar', ax=axes[1], color='lightgreen')
        axes[1].set_title('Confort moyen par type de transport')
        axes[1].set_ylabel('Confort (0-10)')
        axes[1].set_ylim(0, 10)
        
        plt.tight_layout()
        plt.show()
    
    def analyze_popular_routes(self):
        """Analyser les trajets les plus populaires"""
        print("\n" + "=" * 50)
        print("üó∫Ô∏è ANALYSE DES TRAJETS POPULAIRES")
        print("=" * 50)
        
        # Trajets les plus fr√©quents
        routes_populaires = self.df.groupby(['depart', 'arrivee']).size().reset_index(name='count')
        routes_populaires = routes_populaires.sort_values('count', ascending=False)
        
        print("Top 10 des trajets les plus fr√©quents:")
        print(routes_populaires.head(10))
        
        # Stations les plus utilis√©es
        print("\nüîù Stations de d√©part les plus populaires:")
        print(self.df['depart'].value_counts().head(10))
        
        print("\nüîö Stations d'arriv√©e les plus populaires:")
        print(self.df['arrivee'].value_counts().head(10))
    
    def correlation_analysis(self):
        """Analyser les corr√©lations entre variables"""
        print("\n" + "=" * 50)
        print("üìà ANALYSE DES CORR√âLATIONS")
        print("=" * 50)
        
        # Matrice de corr√©lation
        corr_matrix = self.df[['prix', 'rapidite', 'confort']].corr()
        print("Matrice de corr√©lation:")
        print(corr_matrix.round(3))
        
        # Heatmap de corr√©lation
        plt.figure(figsize=(8, 6))
        sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0,
                   square=True, fmt='.3f')
        plt.title('Heatmap des corr√©lations entre prix, rapidit√© et confort')
        plt.tight_layout()
        plt.show()
    
    def comprehensive_analysis(self):
        """Lancer l'analyse compl√®te"""
        self.basic_info()
        self.check_missing_values()
        self.analyze_transport_types()
        self.analyze_prices()
        self.analyze_ratings()
        self.analyze_popular_routes()
        self.correlation_analysis()

# Utilisation
if __name__ == "__main__":
    # Lancer l'analyse
    eda = ExploratoryDataAnalysis("sunuguide_clean_standard.csv")
    eda.comprehensive_analysis()
```

```python
# On nettoie et enrichit le dataset
class DataPreprocessor:
    def __init__(self, df):
        self.df = df.copy()
    
    def clean_data(self):
        # Remplir les valeurs manquantes
        self.df['rapidite'] = self.df['rapidite'].fillna(5.0)  # Valeur moyenne
        self.df['confort'] = self.df['confort'].fillna(5.0)    # Valeur moyenne
        return self.df
    
    def create_features(self):
        # Cat√©goriser les prix
        self.df['categorie_prix'] = pd.cut(self.df['prix'], 
                                         bins=[0, 500, 2000, 50000],
                                         labels=['√©conomique', 'moyen', 'cher'])
        
        # Score global basique
        self.df['score_basique'] = (self.df['rapidite'] + self.df['confort']) / 2
        return self.df
```

```python
class ScoringModel:
    def __init__(self, df):
        self.df = df
        self.max_prix = df['prix'].max()
    
    def calculate_score(self, option, preference='√©quilibr√©'):
        """Calcule un score selon la pr√©f√©rence utilisateur"""
        
        # D√©finir les poids selon la pr√©f√©rence
        if preference == '√©conomique':
            weights = {'prix': 0.7, 'rapidite': 0.2, 'confort': 0.1}
        elif preference == 'rapide':
            weights = {'prix': 0.2, 'rapidite': 0.7, 'confort': 0.1}
        elif preference == 'confortable':
            weights = {'prix': 0.2, 'rapidite': 0.1, 'confort': 0.7}
        else:  # √©quilibr√©
            weights = {'prix': 0.4, 'rapidite': 0.4, 'confort': 0.2}
        
        # Normalisation
        prix_norm = 1 - (option['prix'] / self.max_prix)
        rapidite_norm = option['rapidite'] / 10
        confort_norm = option['confort'] / 10
        
        # Score final
        score = (prix_norm * weights['prix'] + 
                rapidite_norm * weights['rapidite'] + 
                confort_norm * weights['confort'])
        
        return round(score * 10, 2)
```

```python
class SearchEngine:
    def __init__(self, df, scoring_model):
        self.df = df
        self.scoring_model = scoring_model
    
    def find_routes(self, depart, arrivee, preference='√©quilibr√©'):
        """Trouve et score les options de transport"""
        
        # Recherche exacte
        mask = (self.df['depart'].str.lower() == depart.lower()) & \
               (self.df['arrivee'].str.lower() == arrivee.lower())
        
        options = self.df[mask].copy()
        
        if options.empty:
            return None
        
        # Calcul des scores
        options['score'] = options.apply(
            lambda x: self.scoring_model.calculate_score(x, preference), 
            axis=1
        )
        
        # Tri et top 3
        top_3 = options.nlargest(3, 'score')
        return top_3
```

```python
class TransportAssistant:
    def __init__(self, data_path):
        # Chargement et pr√©paration
        df = pd.read_csv(data_path)
        preprocessor = DataPreprocessor(df)
        self.df = preprocessor.clean_data().create_features()
        
        # Initialisation des composants
        self.scoring_model = ScoringModel(self.df)
        self.search_engine = SearchEngine(self.df, self.scoring_model)
    
    def start(self):
        """Lance l'assistant de transport"""
        print("üöó ASSISTANT DE TRANSPORT SUNUGUIDE")
        print("=" * 40)
        
        while True:
            # Saisie utilisateur
            depart = input("\nüìç Point de d√©part: ").strip()
            arrivee = input("üéØ Point d'arriv√©e: ").strip()
            
            print("\nüéõÔ∏è  Pr√©f√©rences de voyage:")
            print("1. √âconomique (prix prioritaire)")
            print("2. Rapide (rapidit√© prioritaire)") 
            print("3. Confortable (confort prioritaire)")
            print("4. √âquilibr√© (mix)")
            
            pref_choice = input("Choisissez (1-4) [4 par d√©faut]: ").strip()
            
            preferences_map = {
                '1': '√©conomique', '2': 'rapide', 
                '3': 'confortable', '4': '√©quilibr√©'
            }
            
            preference = preferences_map.get(pref_choice, '√©quilibr√©')
            
            # Recherche
            recommendations = self.search_engine.find_routes(depart, arrivee, preference)
            
            # Affichage r√©sultats
            self.display_results(recommendations, depart, arrivee, preference)
            
            # Continuer?
            continuer = input("\nüîç Nouvelle recherche? (o/n): ").strip().lower()
            if continuer != 'o':
                print("üëã Bon voyage!")
                break
    
    def display_results(self, recommendations, depart, arrivee, preference):
        """Affiche les recommandations"""
        print(f"\n{'='*50}")
        print(f"üéØ RECHERCHE: {depart} ‚Üí {arrivee}")
        print(f"‚öñÔ∏è  Pr√©f√©rence: {preference}")
        print(f"{'='*50}")
        
        if recommendations is None or len(recommendations) == 0:
            print("‚ùå Aucun trajet direct trouv√©.")
            print("üí° Essayez avec des noms de stations plus g√©n√©riques")
            return
        
        print(f"‚úÖ {len(recommendations)} option(s) trouv√©e(s)\n")
        
        for i, (_, option) in enumerate(recommendations.iterrows(), 1):
            print(f"{i}. üöó {option['type transport']}")
            print(f"   üí∞ Prix: {option['prix']} FCFA")
            print(f"   ‚ö° Rapidit√©: {option['rapidite']}/10")
            print(f"   üõãÔ∏è  Confort: {option['confort']}/10")
            print(f"   ‚≠ê Score: {option['score']}/10")
            print(f"   üìç Trajet: {option['depart']} ‚Üí {option['arrivee']}")
            print()
```
